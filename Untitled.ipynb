````python
from dotenv import load_dotenv
load_dotenv()

from openai import OpenAI
openai_client = OpenAI()
````

````python
def llm(prompt: str):
    response = openai_client.responses.create(
        model="gpt-5.4-mini",
        input=prompt
    )
    return response.output_text
````

## What RAG is

RAG stands for retrieval augmented generation -- we use search to augment what the LLM generates. We want to be able to add relevant context automatically via search instead of providing it manually like what we did above.

It would follow this flow:
```python
def rag(question: string):
    search_results = search(question)
    user_prompt = build_prompt(question, search_results)
    return llm(user_prompt)
```
